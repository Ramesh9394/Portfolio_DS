## NumPy Array Creation

```python
# From list
a = [1,2,3,4]
b = np.array(a)
print(b)
print(type(b))
```

```python
a = np.array([1,22,3,45,6])
print(a)
print(type(a))
```

## Dimensions of Arrays
- 1-D Array
- 2-D Array
- 3-D Array
- Higher Dimensional Array

```python
from numpy.core.fromnumeric import ndim
a = np.array([1,22,3,45,6])
print(a)
print(type(a))
print(a.ndim)
```

```python
a = np.array([[1,2,3,4],[8,9,7,5]])
print(a)
print(type(a))
print(a.ndim)
```

```python
a = np.array([[[1,2,3,4],[8,9,7,5],[3,4,5,9]]])
print(a)
print(type(a))
print(a.ndim)
```

#Zeros

```python
np_zeros = np.zeros(5)
print(np_zeros)
print(np_zeros.ndim)
```

```python
np_zeros_2d = np.zeros((5,5))
print(np_zeros_2d)
print(np_zeros_2d.ndim)
```

```python
np_zeros_3d = np.zeros((3,5,5))
print(np_zeros_3d)
print(np_zeros_3d.ndim)
```

#Ones

```python
np_one = np.ones(10)
print(np_one)
print(np_one.ndim)
```

```python
np_one_2d = np.ones((10,2))
print(np_one_2d)
print(np_one_2d.ndim)
```
